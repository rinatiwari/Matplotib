```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
rainfile_df=pd.read_csv("Resources/avg_rain_state.csv")
rainfile_df
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rainfile_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, rainfile_df["Inches"], color='b', alpha=0.5, align='edge')
plt.xticks(tick_locations, rainfile_df ["State"], rotation='vertical')
```

```python
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0,max(rainfile_df["Inches"])+10)
```

```python
# Set a Title and labels
plt.title("Average Rain per State")
plt.xlabel("State")
plt.ylabel("Rainfall in inches")
```

```python
# Save our graph and show the grap
plt.tight_layout
plt.savefig("Images/Avg_rain_state_RT.png")
plt.show()
```
